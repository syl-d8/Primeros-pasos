```python
import numpy as np
import pandas as pd
import seaborn as sns
import datetime
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('df_initial_32833.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], format='mixed')
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='track_album_release_date', bins=20, kde=True, color='skyblue')
plt.xlabel('Release date')
plt.ylabel('Frequency')
plt.title('Histogram of Release date')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='track_popularity', bins=20, kde=True, color='skyblue')
plt.xlabel('Track Popularity')
plt.ylabel('Frequency')
plt.title('Histogram of Track Popularity')
plt.grid(True)
plt.show()
```

```python
df['release_year'] = df['track_album_release_date'].dt.year
```

```python
df['release_year'].min()
```

```python
df['release_year'].max()
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.scatterplot(data=df, x='release_year', y='track_popularity', color='skyblue', alpha=0.2)
plt.xlabel('Release Year')
plt.ylabel('Track Popularity')
plt.title('Relationship between Release Year and Track Popularity')
plt.grid(True)
plt.show()
```
