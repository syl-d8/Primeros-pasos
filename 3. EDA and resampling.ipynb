```python
# Calculamos la correlaci√≥n de Spearman entre las variables track_popularity y artist_popularity
import scipy.stats
x=df['track_popularity']
y=df['artist_popularity']
r, p = scipy.stats.spearmanr(x, y)
print('R', r)
print('P', p)
```

```python
## Hay correlaci√≥n significativa entre la popularidad del artista y de la canci√≥n, pero principalmente para los valores altos.
## Para balancear los datos de track_popularity hacemos un remuestreo de las canciones con popularidad = 0, manteniendo la distribuci√≥n de popularidad del artista
```

```python
df_pop0= df.loc[df['track_popularity'] == 0]
```

```python
df_pop0.info()
```

```python
df_pop0.sort_values(by=['artist_popularity'], ascending=True, inplace=True)
```

```python
df_pop0.head(5)
```

```python
df_pop0['artist_popularity'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df_pop0, x='artist_popularity', bins=50, kde=True, color='skyblue')
plt.xlabel('Artist Popularity')
plt.ylabel('Frequency')
plt.title('Histogram of Artist Popularity of songs with track_popularity = 0')
plt.grid(True)
plt.show()
```

```python
## Remuestreo para disminuir la categor√≠a de track_popularity = 0

df_sample = df_pop0.sample(n=150, weights='artist_popularity', random_state=42)
```

```python
df_sample['artist_popularity'].value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df_pop0['artist_popularity'], bins=50)
plt.hist(df_sample['artist_popularity'], bins=50)
plt.legend(['Original', 'Sample'])
plt.title('Distribuci√≥n de popularidad de artistas para canciones con popularidad = 0')
plt.show()
```

```python
df2= df.loc[df['track_popularity'] != 0]
len(df2)
```

```python
df= pd.concat([df2, df_sample])
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='track_popularity', bins=100, kde=True, color='skyblue')
plt.xlabel('Track Popularity')
plt.ylabel('Frequency')
plt.title('Histogram of Track Popularity')
plt.grid(True)
plt.show()
```

```python
## Variables num√©ricas
num_col = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'release_year']
```

```python
# Create subplots for histograms
fig, axes = plt.subplots(nrows=len(num_col) // 3 + len(num_col) % 3, ncols=3, figsize=(15, 20))

# Flatten the axes for iteration
axes = axes.flatten()

# Plot histograms for each numerical column
for i, column in enumerate(num_col):
    ax = axes[i]
    ax.hist(df[column], bins=20, alpha=0.7)
    ax.set_title(column)
    ax.set_ylabel('Frequency')
    ax.set_xlabel('Value')
    ax.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
